# Numerical Integration: Trapezoidal Rule
---
## Introduction
The Trapezoidal Rule is a numerical method to approximate definite integrals.

For a function f(x) over [a, b], divide the interval into n equal parts:

$$h = \frac{b-a}{n}$$

The Trapezoidal Rule formula is:

$$\int_a^b f(x) dx \approx \frac{h}{2} \left[f(x_0) + 2\sum_{i=1}^{n-1} f(x_i) + f(x_n)\right]$$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def trapezoidal_rule(f, a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    y = f(x)
    T = (h/2) * (y[0] + 2*np.sum(y[1:n]) + y[n])
    return T
```

## Example 1
Approximate:

$$\int_0^1 e^{-x^2} dx$$

```python
def f(x):
    return np.exp(-x**2)

a = 0
b = 1
n = 6

approx = trapezoidal_rule(f, a, b, n)
approx
```

```python
def plot_trapezoidal(f, a, b, n):
    x = np.linspace(a, b, 1000)
    y = f(x)
    xi = np.linspace(a, b, n+1)
    yi = f(xi)

    plt.figure()
    plt.plot(x, y)
    for i in range(n):
        plt.fill([xi[i], xi[i], xi[i+1], xi[i+1]],
                 [0, yi[i], yi[i+1], 0], alpha=0.3)
    plt.title('Trapezoidal Rule Approximation')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.show()

n=5
plot_trapezoidal(f, a, b, n)
```

## Example 2
Approximate:
\[
\int_1^3 x^2 dx
\]
Exact value = 26/3

```python
def f2(x):
    return x**2

a = 1
b = 3
n = 4

approx2 = trapezoidal_rule(f2, a, b, n)
exact = 26/3

approx2, exact
```

```python
error = abs(exact - approx2)
error
```

## Convergence Study

```python
for n in [2,4,8,16,32]:
    approx = trapezoidal_rule(f2, 1, 3, n)
    print('n =', n, 'Approximation =', approx)
```
