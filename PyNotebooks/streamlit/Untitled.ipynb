Write a Python function to calculate the required sample size for estimating a population mean using $n = \left( \frac{Z\sigma}{E} \right)^2$ where $Z$ is the Z-value corresponding to the confidence level, $\sigma$ is the population standard deviation, and $E$ is the permissible error. Allow the user to input the confidence level, standard deviation, and error.

```python
import numpy as np
import math
import scipy.stats as ss

def sample_size(alfa, sigma, error):
    z_alpha2=ss.norm.ppf(1-alfa/2)
    return np.round((z_alpha2*sigma/error)**2)

sample_size(0.05,1,0.2)
```
